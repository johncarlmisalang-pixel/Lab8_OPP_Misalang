```python
class Book:
    def __init__(self, title, author, is_available=True):
        self.title = str(title)
        self.author = author
        self.is_available = is_available

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author, status):
        is_avail = True if status.lower() == "available" else False
        new_book = Book(title, author, is_avail)
        self.books.append(new_book)

    def display_books(self):
        print("\n--- Library Inventory ---")
        for book in self.books:
            status = "Available" if book.is_available else "Borrowed"
            print(f"Title: {book.title} | Author: {book.author} | Status: {status}")

    def borrow_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                if book.is_available:
                    book.is_available = False
                    return f"Successfully borrowed '{book.title}'."
                else:
                    return f"Error: '{book.title}' is already borrowed."
        return "Error: Book not found."

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                if not book.is_available:
                    book.is_available = True
                    return f"Successfully returned '{book.title}'."
                else:
                    return f"Error: '{book.title}' was not borrowed."
        return "Error: Book not found."

def main():
    library = Library()
    
    try:
        num_books = int(input("How many books to add? "))
    except ValueError:
        print("Invalid input. Please enter a number.")
        return

    for i in range(num_books):
        print(f"\nEntering details for Book {i+1}:")
        title = input("Title: ")
        author = input("Author: ")
        status = input("Status (Available/Borrowed): ")
        library.add_book(title, author, status)

    while True:
        library.display_books()
        print("\nOptions: 1. Borrow  2. Return  3. Exit")
        choice = input("Select an option: ")

        if choice == '1':
            title = input("Enter book title to borrow: ")
            print(library.borrow_book(title))
        elif choice == '2':
            title = input("Enter book title to return: ")
            print(library.return_book(title))
        elif choice == '3':
            break
        else:
            print("Invalid choice.")

if __name__ == "__main__":
    main()
```

```python
class Employee:
    def __init__(self, name, hourly_wage, hours_worked):
        self.name = name
        self.hourly_wage = hourly_wage
        self.hours_worked = hours_worked
        self.base_hours = 160

    def calculate_payroll(self):
        # Determine base and overtime hours
        if self.hours_worked > self.base_hours:
            regular_hours = self.base_hours
            overtime_hours = self.hours_worked - self.base_hours
        else:
            regular_hours = self.hours_worked
            overtime_hours = 0

        # Calculations
        base_salary = regular_hours * self.hourly_wage
        overtime_pay = overtime_hours * (self.hourly_wage * 1.5)
        total_salary = base_salary + overtime_pay

        return {
            "base_salary": base_salary,
            "overtime_pay": overtime_pay,
            "total_salary": total_salary
        }

    def display_statement(self):
        pay = self.calculate_payroll()
        print(f"\n--- Pay Slip: {self.name} ---")
        print(f"Total Hours Worked: {self.hours_worked}")
        print(f"Base Salary:       ${pay['base_salary']:.2f}")
        print(f"Overtime Pay:      ${pay['overtime_pay']:.2f}")
        print(f"Total Net Pay:     ${pay['total_salary']:.2f}")


def get_positive_float(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value < 0:
                print("Error: Value cannot be negative.")
                continue
            return value
        except ValueError:
            print("Error: Please enter a valid numerical value.")


def main():
    print("Welcome to the Employee Payroll System")
    
    # Ensure at least one employee is entered
    while True:
        try:
            num_employees = int(input("Enter the number of employees: "))
            if num_employees <= 0:
                print("Error: You must enter at least 1 employee.")
                continue
            break
        except ValueError:
            print("Error: Please enter a whole number.")

    employees = []

    for i in range(num_employees):
        print(f"\nData for Employee #{i + 1}:")
        name = input("Enter Name: ").strip()
        while not name:
            name = input("Name cannot be empty. Enter Name: ").strip()
            
        wage = get_positive_float(f"Enter Hourly Wage for {name}: ")
        hours = get_positive_float(f"Enter Total Hours Worked for {name}: ")
        
        employees.append(Employee(name, wage, hours))

    print("\n" + "="*30)
    print("      MONTHLY PAYROLL SUMMARY")
    print("="*30)
    
    for emp in employees:
        emp.display_statement()

if __name__ == "__main__":
    main()
```

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def __str__(self):
        return f"{self.name} - ${self.price} ({self.stock} in stock)"

class ShoppingCart:
    def __init__(self):
        self.items = {}  # Format: {product_name: [price, quantity]}
        self.total_price = 0.0

    def add_item(self, product, quantity):
        if quantity <= 0:
            return "Quantity must be greater than zero."
        
        if quantity > product.stock:
            return f"Error: Only {product.stock} units of {product.name} available."

        # Update product stock
        product.stock -= quantity
        
        # Add to cart logic
        if product.name in self.items:
            self.items[product.name][1] += quantity
        else:
            self.items[product.name] = [product.price, quantity]
        
        self.calculate_total()
        return f"Added {quantity} x {product.name} to your cart."

    def remove_item(self, product_name, inventory):
        if product_name not in self.items:
            return f"Error: {product_name} is not in your cart."
        
        # Return stock to inventory
        removed_qty = self.items[product_name][1]
        for prod in inventory:
            if prod.name == product_name:
                prod.stock += removed_qty
                break
        
        del self.items[product_name]
        self.calculate_total()
        return f"Removed {product_name} from your cart."

    def calculate_total(self):
        self.total_price = sum(details[0] * details[1] for details in self.items.values())

    def view_cart(self):
        if not self.items:
            return "\nYour cart is empty."
        
        display = "\n--- Your Shopping Cart ---"
        for name, details in self.items.items():
            display += f"\n{name}: {details[1]} x ${details[0]} = ${details[0] * details[1]:.2f}"
        display += f"\nTotal Price: ${self.total_price:.2f}"
        return display

def main():
    # Predefined Inventory
    inventory = [
        Product("Laptop", 1200, 5),
        Product("Mouse", 25, 10),
        Product("Keyboard", 50, 8),
        Product("Monitor", 200, 3)
    ]
    
    cart = ShoppingCart()

    while True:
        print("\n--- Available Products ---")
        for i, prod in enumerate(inventory):
            print(f"{i + 1}. {prod}")
        
        print("\nOptions: 1. Add Item  2. Remove Item  3. View Cart  4. Checkout")
        choice = input("Select an option: ")

        if choice == '1':
            try:
                idx = int(input("Enter product number: ")) - 1
                if 0 <= idx < len(inventory):
                    qty = int(input(f"How many {inventory[idx].name}s? "))
                    print(cart.add_item(inventory[idx], qty))
                else:
                    print("Invalid product selection.")
            except ValueError:
                print("Invalid input. Please enter numbers only.")

        elif choice == '2':
            if not cart.items:
                print("Cart is empty.")
                continue
            name = input("Enter the name of the product to remove: ")
            # Using title() to match case-sensitivity if necessary
            print(cart.remove_item(name, inventory))

        elif choice == '3':
            print(cart.view_cart())

        elif choice == '4':
            print(cart.view_cart())
            print("Thank you for shopping! Proceeding to checkout...")
            break
        else:
            print("Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```
