```python
import numpy as np
import pandas as pd

# Basic Info
STUDENT_NAME = "John Carl Misalang"
STUDENT_ID = 20250478

# 1. Read the dataset
df = pd.read_csv('Most Popular Programming Languages.csv')

# --- FIX: Print columns to debug if 'Date' is missing ---
print("Detected Columns:", df.columns.tolist())

# Check if 'Date' exists; if not, look for 'Month' or 'date'
if 'Date' not in df.columns:
    if 'Month' in df.columns:
        df.rename(columns={'Month': 'Date'}, inplace=True)
    elif 'date' in df.columns:
        df.rename(columns={'date': 'Date'}, inplace=True)

# 2. Convert date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# 3. Sort language columns alphabetically (excluding the 'Date' column)
languages = sorted([col for col in df.columns if col != 'Date'])

# 4. Assign language based on the last three digits of Student ID
language_index = (STUDENT_ID % 1000) % len(languages)
assigned_language = languages[language_index]

# 5. Extract data for the assigned language
lang_df = df[['Date', assigned_language]].copy()
lang_df.columns = ['Month', 'Popularity']

# 6. Compute Growth Rate (percentage)
lang_df['Growth_Rate'] = lang_df['Popularity'].pct_change() * 100

# 7. Calculate Rolling Statistics (12-month window)
lang_df['Moving_Avg'] = lang_df['Popularity'].rolling(window=12).mean()
lang_df['Moving_STD'] = lang_df['Popularity'].rolling(window=12).std()

# 8. Classify Growth Phases
growth_threshold = 5.0
decline_threshold = -5.0

lang_df['Phase'] = np.select(
    [
        lang_df['Growth_Rate'] > growth_threshold,
        lang_df['Growth_Rate'] < decline_threshold
    ],
    ['Growth', 'Decline'],
    default='Stable'
)

# 9. Key Performance Metrics
initial_val = lang_df['Popularity'].iloc[0]
final_val = lang_df['Popularity'].iloc[-1]
overall_growth = ((final_val - initial_val) / initial_val) * 100
phase_counts = lang_df['Phase'].value_counts()

# --- OUTPUTS ---
print(f"\nStudent Name: {STUDENT_NAME}")
print(f"Assigned Language: {assigned_language}")
print("-" * 30)
print("\nDataFrame Output (First 10 records):")
print(lang_df.head(10))
print("\nStatistical Summary Table:")
print(lang_df['Popularity'].describe())
print("\nGrowth Phase Counts:")
print(phase_counts)
print("\nKey Performance Metrics:")
print(f"Initial Popularity: {initial_val:.2f}")
print(f"Final Popularity: {final_val:.2f}")
print(f"Overall Growth Percentage: {overall_growth:.2f}%")
```

```python
import numpy as np
import pandas as pd

# Basic Info
STUDENT_NAME = "John Carl Misalang"
STUDENT_ID = 20250478

# 1. Load the dataset
# Update filename if your actual file is named differently
df = pd.read_csv('Most Popular Programming Languages.csv')

# Handle column naming variations from Kaggle (convert Date to Month if needed)
if 'Date' in df.columns:
    df.rename(columns={'Date': 'Month'}, inplace=True)
elif 'date' in df.columns:
    df.rename(columns={'date': 'Month'}, inplace=True)

# Parse the Month column as datetime type
df['Month'] = pd.to_datetime(df['Month'])

# 2. Select assigned language for individual analysis
# Sort alphabetically (excluding the 'Month' column)
languages = sorted([col for col in df.columns if col != 'Month'])
language_index = (STUDENT_ID % 1000) % len(languages)
assigned_language = languages[language_index]

# Extract data
lang_df = df[['Month', assigned_language]].copy()
lang_df.columns = ['Month', 'Popularity']

# 3. Compute growth rate
lang_df['Growth_Rate'] = lang_df['Popularity'].pct_change() * 100

# 4. Smooth the trend
lang_df['Moving_Avg'] = lang_df['Popularity'].rolling(window=6).mean()
lang_df['Moving_STD'] = lang_df['Popularity'].rolling(window=6).std()

# 5. Analyze central tendency using NumPy
# Drop NaN values (like the first row) before calculating mean and std
clean_growth = lang_df['Growth_Rate'].dropna()
overall_mean = np.mean(clean_growth)
overall_std = np.std(clean_growth)

# 6. Apply lifecycle classification using np.select()
conditions = [
    np.abs(lang_df['Growth_Rate']) <= 1.0,                                      # Maturity: |Growth rate| <= 1%
    (lang_df['Growth_Rate'] < 0) & (lang_df['Growth_Rate'] < -overall_std),     # Decline: < 0 and below (-1 * std)
    lang_df['Growth_Rate'] > overall_mean,                                      # Growth: > mean
    (lang_df['Growth_Rate'] > 0) & (lang_df['Growth_Rate'] <= overall_mean)     # Introduction: > 0 but below mean
]

choices = ['Maturity', 'Decline', 'Growth', 'Introduction']

# Assign phases (Using 'Transition' for any edge-case months that fall between thresholds)
lang_df['Lifecycle_Phase'] = np.select(conditions, choices, default='Transition')

# Clean up the very first month which has no prior data to compare to
lang_df.loc[lang_df['Growth_Rate'].isna(), 'Lifecycle_Phase'] = 'Initial'

# 7. Summarize distribution
# Exclude the 'Initial' row for accurate calculations
valid_phases = lang_df[lang_df['Lifecycle_Phase'] != 'Initial']['Lifecycle_Phase']
phase_counts = valid_phases.value_counts()
phase_percentages = (phase_counts / len(valid_phases)) * 100

# Identify dominant stage
dominant_stage = phase_counts.idxmax()

# --- OUTPUTS ---
print(f"Student Name: {STUDENT_NAME}")
print(f"Assigned Language: {assigned_language}")
print("-" * 60)

print("\nLifecycle Classification Table (Last 10 Months):")
print(lang_df.tail(10))

print("-" * 60)
print("\nSummary Output:")
print("\nCount per Lifecycle Phase:")
print(phase_counts)
print("\nPercentage per Lifecycle Phase (%):")
print(phase_percentages.round(2))

print(f"\nDominant Stage:")
print(f"-> {assigned_language} is primarily categorized by the '{dominant_stage}' phase.")
if dominant_stage == 'Maturity':
    print(f"   Interpretation: {assigned_language} is currently in the maturity phase with stable global adoption.")
elif dominant_stage == 'Growth':
    print(f"   Interpretation: {assigned_language} is experiencing periods of significant, continuous adoption.")
elif dominant_stage == 'Decline':
    print(f"   Interpretation: {assigned_language} has faced consistent drops in popularity compared to its historical average.")
elif dominant_stage == 'Introduction':
    print(f"   Interpretation: {assigned_language} sees slow but steady positive adoption.")

print("-" * 60)
print("\nInterpretive Insights:")

# Check transitions over the last 12 months for resurgence/decline insights
recent_12_months = lang_df['Lifecycle_Phase'].tail(12).tolist()
# Reduce consecutive duplicates to show clear transitions
transitions = pd.Series(recent_12_months).loc[pd.Series(recent_12_months).shift() != pd.Series(recent_12_months)].tolist()

print(f"Transitions over the last year: {' -> '.join(transitions)}")

if 'Decline' in recent_12_months[-3:]:
    print("Trajectory: Indications of a recent decline in popularity.")
elif 'Growth' in recent_12_months[-3:]:
    print("Trajectory: Indications of a possible resurgence and strong recent growth.")
elif 'Maturity' in recent_12_months[-3:]:
    print("Trajectory: The language's adoption rate has flattened into stable maturity recently.")
```

```python
import numpy as np
import pandas as pd

# Basic Info
STUDENT_NAME = "John Carl Misalang"
STUDENT_ID = 20250478

# 1. Load dataset
# Update filename if your actual file is named differently
df = pd.read_csv('Most Popular Programming Languages.csv')

# Handle column naming variations from Kaggle
if 'Date' in df.columns:
    df.rename(columns={'Date': 'Month'}, inplace=True)
elif 'date' in df.columns:
    df.rename(columns={'date': 'Month'}, inplace=True)

# Parse the Month column as datetime type
df['Month'] = pd.to_datetime(df['Month'])

# 2. Select assigned languages
languages = sorted([col for col in df.columns if col != 'Month'])
lang1_index = (STUDENT_ID % 1000) % len(languages)
# Next language alphabetically (using modulo to wrap around if it's the last one)
lang2_index = (lang1_index + 1) % len(languages) 

lang1 = languages[lang1_index]
lang2 = languages[lang2_index]

comp_df = df[['Month', lang1, lang2]].copy()

# 3. Compute central tendencies
mean_1, median_1, std_1 = comp_df[lang1].mean(), comp_df[lang1].median(), comp_df[lang1].std()
mean_2, median_2, std_2 = comp_df[lang2].mean(), comp_df[lang2].median(), comp_df[lang2].std()

# 4. Normalize data using NumPy (Z-Score normalization to align scales)
comp_df[f'{lang1}_Norm'] = (comp_df[lang1] - mean_1) / std_1
comp_df[f'{lang2}_Norm'] = (comp_df[lang2] - mean_2) / std_2

# 5. Calculate correlation coefficient
correlation = np.corrcoef(comp_df[lang1], comp_df[lang2])[0, 1]

# 6. Determine dominance ratio
total_months = len(comp_df)
dom_1_count = np.sum(comp_df[lang1] > comp_df[lang2])
dom_2_count = np.sum(comp_df[lang2] > comp_df[lang1])

dom_ratio_1 = (dom_1_count / total_months) * 100
dom_ratio_2 = (dom_2_count / total_months) * 100

# 7. Compute Relative Performance Index (RPI) 
# RPI = Ratio of Means (Lang A vs Lang B)
rpi = mean_1 / mean_2 if mean_2 != 0 else np.nan

# 8. Identify cross-over points (months when the leader changes)
# Create a series of signs (1 if Lang1 > Lang2, -1 if Lang2 > Lang1, 0 if equal)
diff = comp_df[lang1] - comp_df[lang2]
signs = np.sign(diff)
# Find indices where the sign changes
sign_changes = np.where(np.diff(signs) != 0)[0] + 1 
crossovers = comp_df.iloc[sign_changes][['Month', lang1, lang2]]

# 9. Summarize comparative statistics in a DataFrame
stats_summary = pd.DataFrame({
    'Metric': ['Mean', 'Median', 'Std Dev', 'Dominance Ratio (%)'],
    lang1: [mean_1, median_1, std_1, dom_ratio_1],
    lang2: [mean_2, median_2, std_2, dom_ratio_2]
})

# --- OUTPUTS ---
print(f"Student Name: {STUDENT_NAME}")
print(f"Assigned Language 1: {lang1}")
print(f"Assigned Language 2: {lang2}")
print("-" * 60)

print("\nComparative Statistics Table:")
print(stats_summary.to_string(index=False))

print(f"\nCorrelation between {lang1} and {lang2}: {correlation:.4f}")
print(f"Relative Performance Index (RPI: {lang1}/{lang2}): {rpi:.4f}")

print("\nCross-Over Points (Months where the dominant language swapped):")
if crossovers.empty:
    print("No cross-over points found. One language dominated entirely.")
else:
    # Display the first few crossovers if there are many
    print(crossovers.head(10).to_string(index=False))
    if len(crossovers) > 10:
        print(f"... and {len(crossovers) - 10} more cross-over events.")

print("-" * 60)
print("\nInterpretation:")
# Determine the more consistent/resilient language based on Std Dev and Dominance
dominant = lang1 if dom_ratio_1 > dom_ratio_2 else lang2
consistent = lang1 if std_1 < std_2 else lang2

print(f"-> Dominance: {dominant} has been more historically dominant, leading for {max(dom_ratio_1, dom_ratio_2):.1f}% of the recorded months.")

if correlation > 0.7:
    print(f"-> Correlation: Strong positive correlation ({correlation:.2f}). Both languages tend to grow or decline together.")
elif correlation < -0.7:
    print(f"-> Correlation: Strong negative correlation ({correlation:.2f}). As one language grows, the other tends to decline.")
else:
    print(f"-> Correlation: Weak or moderate correlation ({correlation:.2f}). Their popularity trends are relatively independent.")

print(f"-> Consistency: {consistent} shows less variability (lower Standard Deviation), indicating more resilient, stable adoption over the 20-year span.")
```
